### 6. Splitting the dataset into train and test and applying Machine Learning Models

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split (pixels,labels)
```

### 7. Performing Principal Component Analysis

```python
#PCA step
from sklearn.decomposition import PCA
pca=PCA(135).fit(x_train)
```

```python
per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
labels = [str(x) for x in range(1,len(per_var)+1)]
 
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.xticks(rotation=90)
plt.title('Plot of PCA based results',size=21)
plt.show()
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('Principal Components - Variance Ratio\n',size=22)
plt.show()
```

```python
def show_eigenfaces(pca):
   #Displaying Eigenfaces
   fig, axes = plt.subplots(3, 8, figsize=(15, 7),subplot_kw={'xticks':[], 'yticks':[]})
   for i, ax in enumerate(axes.flat):
       ax.imshow(pca.components_[i].reshape(64, 64), cmap='gray')
       ax.set_title("PC " + str(i+1))
   plt.show()

show_eigenfaces(pca)
```

### 8. Classifying images using SVM

```python
x_train_pca=pca.transform(x_train)#create_classifer
from time import time
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
clf=SVC(kernel="rbf",C=1000,gamma=0.01)
clf.fit(x_train_pca,y_train)
```

### 9. Testing results

```python
#testing
x_test_pca=pca.transform(x_test)
y_pred=clf.predict(x_test_pca)

print(classification_report(y_test,y_pred))
```

```python
import pixellib
from pixellib.tune_bg import alter_bg
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from IPython.display import Image as img
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10, 10
```

```python
change_bg = alter_bg()
```

```python
dir(change_bg)
```

```python
change_bg.load_pascalvoc_model("deeplabv3_xception_tf_dim_ordering_tf_kernels.h5")
```

```python
file_name = "/content/img.jpg"
bg_file_name = "/content/background_img.jpg"
```

```python
plt.imshow(Image.open(file_name))
plt.imshow(Image.open(bg_file_name))
```

```python
change_bg.blur_bg(file_name, 
                  moderate = True , 
                  output_image_name="blur_img.jpg")
```

```python

plt.imshow(Image.open("blur_img.jpg"))
```

```python
change_bg.gray_bg(file_name,output_image_name="gray_img.jpg")
plt.imshow(Image.open("gray_img.jpg"))
```

```python
change_bg.color_bg(file_name,colors = (123, 123, 123),output_image_name="colored_bg.jpg")
plt.imshow(Image.open("colored_bg.jpg"))
```

```python
change_bg.change_bg_img(f_image_path = file_name,b_image_path = bg_file_name,output_image_name="new_img.jpg")
plt.imshow(Image.open("new_img.jpg"))
```

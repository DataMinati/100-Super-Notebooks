```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

year_store_sell_df = df.groupby(['store', 'year'])['num_sold'].sum().unstack().T.reset_index()
```

```python
heatmap1_data = pd.pivot_table(df, values='num_sold', index=['store'], columns='year')

plt.subplots(figsize = (12,4), dpi = 80)
sns.heatmap(heatmap1_data, cmap="Purples");
```

```python
fig, ax = plt.subplots(figsize = (12,6), dpi = 80)

ax = sns.barplot(y=year_store_sell_df['KaggleMart'], x=year_store_sell_df['year'],
              palette=['#ffb74b', '#ffd6a4', '#ceb1f2', '#a071ff', '#6f71f7'])

ax1 = ax.twiny()
sns.lineplot(data=year_store_sell_df, marker='o', x='year', y='KaggleMart', color='#3126ff', linewidth = 0.7)

for i in ax.patches:
    ax.text(i.get_x()+0.15, i.get_height()-150000, \
            '{:,d}'.format(int(i.get_height())), fontsize=15,weight='bold',
                color='white')
    
    
plt.title('\nSells in KaggleMart Store from 2015 to 2018\n', size=20, color='grey')
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=12)
ax.set_xlabel('\nYear\n', fontsize=15, color='grey')
ax.set_ylabel('\nNumber of Sell\n', fontsize=15, color='grey')
ax.axes.set_ylim(-20000, 1200000)


ax1.axes.xaxis.set_visible(False)
for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
    
for i in ['top', 'right', 'bottom']:
    ax1.spines[i].set_visible(False)

ax.spines['bottom'].set_bounds(0.005, 3)
ax.spines['left'].set_bounds(0, 1200000)
```

```python
fig, ax = plt.subplots(figsize = (12,6), dpi = 80)

ax = sns.barplot(y=year_store_sell_df['KaggleRama'], x=year_store_sell_df['year'],
              palette=['#cd7cf2', '#f27ec8', '#fa70b3', '#ff5e86', '#ff1732'])

ax1 = ax.twiny()
sns.lineplot(data=year_store_sell_df, marker='o', x='year', y='KaggleRama', color='black', linewidth = 0.7)

for i in ax.patches:
    ax.text(i.get_x()+0.15, i.get_height()-150000, \
            '{:,d}'.format(int(i.get_height())), fontsize=15,weight='bold',
                color='white')
    
    
plt.title('\nSells in KaggleRama Store from 2015 to 2018\n', size=20, color='grey')
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=12)
ax.set_xlabel('\nYear\n', fontsize=15, color='grey')
ax.set_ylabel('\nNumber of Sell\n', fontsize=15, color='grey')
ax.axes.set_ylim(-30000, 2000000)


ax1.axes.xaxis.set_visible(False)
for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
    
for i in ['top', 'right', 'bottom']:
    ax1.spines[i].set_visible(False)

ax.spines['bottom'].set_bounds(0.005, 3)
ax.spines['left'].set_bounds(0, 2000000)
```

```python
df['month'] = df['date'].dt.month

month_names = list(calendar.month_name)[1:]

month_store_sell_df = df.groupby(['store', 'month'])['num_sold'].sum().unstack().T.reset_index()
month_store_sell_df.month = month_names
```

```python
heatmap2_data = pd.pivot_table(df, values='num_sold', index=['store'], columns='month')
heatmap2_data.columns = month_names

plt.subplots(figsize = (12,4), dpi = 80)
sns.heatmap(heatmap2_data, cmap="YlGnBu");
```

```python
fig, ax = plt.subplots(figsize = (12.5,5), dpi = 80)

ax=sns.barplot(y=month_store_sell_df['KaggleMart'], x=month_store_sell_df['month'], palette='rainbow')
ax1 = ax.twinx()
sns.lineplot(data=month_store_sell_df, marker='o', x='month', y='KaggleMart', color = 'white', alpha = .8)

total = sum(month_store_sell_df['KaggleMart'])
for i in ax.patches:
    ax.text(i.get_x()+0.06, i.get_height()-20000, \
            str(round((i.get_height()*100)/total, 2))+'%', fontsize=11,weight='bold',color='white')

plt.title('\nSell in KaggleMart over the Year (2015-2018)\n', size=20, color='grey')
ax.axes.set_ylim(-10000, 400000)
ax1.axes.set_ylim(-10000, 500000)

ax1.axes.yaxis.set_visible(False)
ax.set_xlabel('\nMonths\n', fontsize=15, color='grey')
ax.set_ylabel('\nNumber of Sell\n', fontsize=15, color='grey')

for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
    ax1.spines[i].set_visible(False)

for i in  [ax, ax1]:
    i.spines['bottom'].set_bounds(0.005, 11)
    i.spines['left'].set_bounds(0, 400000)
```

```python
fig, ax = plt.subplots(figsize = (12.5,5), dpi = 80)

ax=sns.barplot(y=month_store_sell_df['KaggleRama'], x=month_store_sell_df['month'], palette='cool')
ax1 = ax.twinx()
sns.lineplot(data=month_store_sell_df, marker='o', x='month', y='KaggleRama', color = 'white', alpha = .8)

total = sum(month_store_sell_df['KaggleRama'])
for i in ax.patches:
    ax.text(i.get_x()+0.06, i.get_height()-50000, \
            str(round((i.get_height()*100)/total, 2))+'%', fontsize=11,weight='bold',color='white')

plt.title('\nSell in KaggleRama over the Year (2015-2018)\n', size=20, color='grey')
ax.axes.set_ylim(-10000, 800000)
ax1.axes.set_ylim(-10000, 1000000)

ax1.axes.yaxis.set_visible(False)
ax.set_xlabel('\nMonths\n', fontsize=15, color='grey')
ax.set_ylabel('\nNumber of Sell\n', fontsize=15, color='grey')

for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
    ax1.spines[i].set_visible(False)

for i in  [ax, ax1]:
    i.spines['bottom'].set_bounds(0.005, 11)
    i.spines['left'].set_bounds(0, 800000)
```
